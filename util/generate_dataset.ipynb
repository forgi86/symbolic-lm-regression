```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
x = np.linspace(0, 10, 200)

true_amplitude = 2.5
true_frequency = 1.2
y_clean = true_amplitude * np.sin(true_frequency * x)

# Adding Gaussian Noise (Standard Deviation = 0.4)
noise = np.random.normal(0, 0.4, x.size)
y_noisy = y_clean + noise
```

```python
df = pd.DataFrame({'x': x, 'y': y_noisy})
df.to_csv('data.csv', index=False)
```

```python
plt.plot(x, y_clean, label='True Function (Clean)', color='blue')
plt.plot(x, y_noisy, label='Noisy Data', color='red')
```
